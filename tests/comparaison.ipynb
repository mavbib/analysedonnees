```python
documentsMesh3 = []

from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()

for sen3 in range(0, len(filteredMesh3)):
    # caractere speciaux
    document3 = re.sub(r'\W', ' ', str(filteredMesh3[sen3]))
    
    # carateres solo
    document3 = re.sub(r'\s+[a-zA-Z]\s+', ' ', document3)
    
    # caractere de depart
    document3 = re.sub(r'\^[a-zA-Z]\s+', ' ', document3) 
    
    #espace multi pour un seul espace
    document3 = re.sub(r'\s+', ' ', document3, flags=re.I)
    
    # prefixe b\
    #document = re.sub(r'^b\s+', '', document)
    
    # Lowercase celui du TF-IDF fait des erreurs selon le format des donnees
    document3 = document3.lower()
    
    # lemmatization lemminer?
    document3 = document3.split()
     
    document3 = [stemmer.lemmatize(word) for word in document3]
    document3 = ' '.join(document3)
    
    documentsMesh3.append(document3)
    documentsMesh3 = [i for i in documentsMesh3 if i]
```
